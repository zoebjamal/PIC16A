### PageRank

What is the most important website on the internet? Who is the "key player" on a sports team? Which countries are the most central players in the world economy? There is no one correct answer to any of these questions, but there is a most profitable one. [PageRank](https://en.wikipedia.org/wiki/PageRank) is an algorithm for ranking individual elements of complex systems, invited by Sergey Brin and Larry Page. It was the first and most famous algorithm used by the Google Search engine, and it is fair to say that the internet as we know it today would not exist without PageRank. 

In this assignment, we will implement PageRank. There are many good ways to implement this algorithm, but in this assignment we will use our newfound skills with object-oriented programming and iterators. 

### How it works

For the purposes of this example, let's assume that we are talking about webpages. PageRank works by allowing a "random surfer" to move around webpages by following links. Each time the surfer lands on a page, it then looks for all the links on that page. It then picks one at random and follows it, thereby arriving at the next page, where the process repeats. Eventually, the surfer will visit all the pages one or more times. Pages that the surfer visits more frequently have higher *PageRank scores*. Because the surfer moves between linked pages, PageRank expresses an intuitive idea: **important pages are linked to other important pages.** [This diagram](https://en.wikipedia.org/wiki/PageRank#/media/File:PageRanks-Example.jpg) from Wikipedia gives a nice illustration. Note that more important webpages (higher PageRank) tend to be connected to other important webpages.

#### Your Solution
I will define a function `data_to_dictionary`
- First I will initialize an empty dict `d`
- I will then loop through the length of `data` and make the first element of each tuple in `data` into a key, whose value is an empty list. 
- Next, I will iterate throught the length of `data` again, checking if the first element in each tuple is present in `d`, by seeing if it is `in list(d.keys())`
- If it is, I will get the value for that key (a list) and append the second element in `data[i]`

````python
def data_to_dictionary(data):
    """
    This function returns a dictionary given a data set that is a list of tuples. The dict's keys are the
    unique mentioners of the tuple and the value is a list of mentioned people that correspond to that mentioner
    
    Parameters
    ----------
    data: user supplied data set that is a list of tuples
    
    Returns
    -------
    d: dictionary version of the user supplied data set
    """
    d = {} # initializing empty dict
    for i in range(len(data)): # looping through length of data
        d[data[i][0]] = [] # setting each unique mentioner to an empty list
    for i in range(len(data)): # looping through the length of data again
        if data[i][0] in list(d.keys()): # checking to see if data[i][0] is in the list of keys in dict d
            d[data[i][0]].append(data[i][1]) # if it is, append the second element of that tuple to the list corresponding
                                             # to that key
    return d

data_to_dictionary(data)
````

We are going to use iteration to implement the PageRank algorithm. This  means we are going to imagine a surfer who is following the links in our data set. **Implement the following two methods:**

#### Your Solution
For this problem, I will first declare the `PR_Iterator()` class and construct the required methods
- `__init__()` will set `i` to `0` and the `current_state` to `hamilton`
- `follow_link()` set the `next_state` to a random element in the values of `link_dict[self.current_state]`. If this value is the same as `current_state`, I will call the `teleport` function to set it to a random key in the `link_dict` dictionary
- `teleport()` sets the `current_state` to a random key in the `link_dict` dictionary
- `__next__()` will first check if the `iteration_limit` has been reached (if it has it will raise a `StopIterration`). If not, it will call `follow_link()` 85% of the time and `teleport()` the other 15% of the time. Finally, it will increment `i` by `1` and return `current_state`

````python
import random
class PR_Iterator():
    """
    A class that can iterate through the PR_DiGraph class.
    
    Returns
    -------
    current_state: the person that we have 'surfed' to
    """
    def __init__(self, D):
        """
        initialize PR_Iterator object
        
        Parameters
        ----------
        self: the object
        D: PR_DiGraph object
        """
        # print("PR_Iterator object created")
        self.i = 0
        self.current_state = 'hamilton' # setting the instance variables
    
    def follow_link(self, D):
        """
        Pick a link to go to. If not the same as current, set as current. If it is current, teleport to a new option
        
        Parameters
        ----------
        self: the object
        D: PR_DiGraph object
        """
        try: # try except clause to avoid KeyError
            next_state = random.choice(D.link_dict[self.current_state]) # setting next_state to a random element in value list
            # print(self.current_state, next_state)
            if next_state != self.current_state: # checking to see if next_state can be used or if we need to be teleported
                self.current_state = next_state
                # print(self.current_state)
            elif next_state == self.current_state:
                # print("need to teleport")
                self.teleport(D)
        except KeyError:
            self.teleport(D) # if there is a KeyError, we will teleport
   
    def teleport(self, D):
        """
        Sets current_state to a random key in link_dict
        
        Parameters
        ----------
        self: the object
        D: PR_DiGraph object
        """
        self.current_state = random.choice(list(D.link_dict.keys())) # setting current_state to a random key in link_dict
        # print(self.current_state)
        
    def __next__(self):
        """
        Iterates through the PR_DiGraph object
        
        Parameters
        ----------
        self: the object
        
        Return
        ------
        self.current_state: the person we are currently looking at
        """
        # print("PR_Iterator.__next__() called")
        if self.i == D.iteration_limit: # checking to see if iteration_limit has been reached
            # print("iteration_limit reached")
            raise StopIteration
        elif random.random() < .85: # calls follow_link 85% of the time
            self.follow_link(D)
        else: # calls teleport 15% of the time
            self.teleport(D)
        self.i += 1 # increments self.i
        return self.current_state
````
